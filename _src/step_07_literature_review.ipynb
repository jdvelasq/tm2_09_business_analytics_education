```python
from techminer2.documents import document_metrics  # type: ignore
from techminer2.documents import select_documents  # type: ignore

frame = document_metrics(
    #
    # DATABASE PARAMS:
    root_dir="../",
    database="main",
    year_filter=(None, None),
    cited_by_filter=(None, None),
    sort_by="global_cited_by_highest",
)

frame = frame[frame.document_type == "Review"]
frame = frame.reset_index(drop=True)
frame.to_csv("../reports/sec_20_literature_review/reviews.csv", encoding="utf-8")
titles = frame.raw_document_title.to_list()

documents = select_documents(
    #
    # DATABASE PARAMS:
    root_dir="../",
    database="main",
    year_filter=(None, None),
    cited_by_filter=(None, None),
    sort_by="global_cited_by_highest",
    raw_document_title=titles,
)


documents = "\n--\n\n".join(documents)

with open(
    "../reports/sec_20_literature_review/reviews.txt", "wt", encoding="utf-8"
) as f:
    print(documents, file=f)


display(frame)
```
